# 连续型随机变量的函数的分布
\(Y = g(X)\)
F_Y(y) = P(Y<=y) = P(g(X)<=y) = 对应的积分
先通过上面的方法求得分布函数，然后求导以后可以得到概率密度函数


```python
import sympy as sp
import base
import sympy as sp

# 定义符号变量
x, y = sp.symbols('x y')

# 定义待积分的分段函数 f_x(x)
fx = sp.Piecewise(
    (1/20, sp.And(x >= 0, x <= 20)),  # 当 0 <= x <= 20 时，函数值为 1/20
    (0, True)  # 否则为 0
)

# 定义积分的上限函数 g(x)，此处以 g(x) <= y 为例
gx = 5 * x

def get_F_Y(fx, gx, x):
    # 设置积分上下限，仍然保持 g(x) <= y 的隐式条件
    lower_limit = -sp.oo  # 负无穷
    upper_limit = sp.solve(gx - y, x)  # 隐式定义上限 g(x) <= y
    # 计算积分 (上限为隐式的 g(x) <= y)
    integral = sp.integrate(fx, (x, lower_limit, upper_limit))    
    return integral

    # 显示结果
integral = get_F_Y(fx, gx, x)
sp.pprint(integral)

base.my_plot(integral, y, -30, 130)
```
