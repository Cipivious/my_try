# 统计的基本概念

1. 总体 随机变量X所有可能取值的集合
2. 个体 指总体中某个具体的元素
3. 总体容量 指总体中可能取值的数量的总和，可能是无限的
4. 样本 从总体X当中随机的选出多个个体组成的集合叫做样本，个体的数量称为样本容量，它们的值称为样本观测值，一般来说，样本的概念与实际的调研是相联系的。
5. 统计量 它是样本观测值集合的函数，对于一组具体的样本，统计量会有具体的值称为统计量的观测值，以下是一些常见的统计量：
    - 样本均值
    - 样本方差 样本方差的期望值要和随机变量的方差相同，所以除（n-1）
    这里要注意，样本均值和样本方差是统计量，将具体的样本的观测值带入公式可以得到对应统计量的观测值
    - 样本矩（原点，中心）
    - 次序统计量（最大值，最小值，极差，中位数，p-分位数，经验分布函数）

6. 样本的统计量也服从一定的分布，可以看作是随机变量的函数的分布，我们取样的时候一般都认为是独立同分布的简单随机抽样，这样就可以方便的计算出统计量所符合的分布，而我们把通过理论计算出的统计量的分布，和实际观测值代入统计量函数得到的结果进行比较，就可以得到参数可能的取值或者区间，或者其它一些有用的信息。

    以样本均值为例解释一下，比如X服从概率分布f（X），做n次简单随机抽样得到X1，X2...Xn，然后计算
    E(X1 + X2 + ... + Xn) = E(X1) + E(X2) + ... + E(Xn) = n*E(X) 得
    E((X1 + X2 + ... + Xn)/n) = E(X)
    这就是统计量 （X1 + X2 + ... + Xn）/ n， 也就是一般意义上的均值，他所具有的特征，这时如果我们通过一些手段计算出原有随机变量的数学期望，其中可能带有一些未知参数，然后我们就可以和实际的测量结果进行比较估计参数的值。
